```python
# 98. Create 4 subplots: histogram, bar, box, scatter.
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
sns.histplot(df['age'].dropna(), bins=20, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Age Distribution (Histogram)')
sns.countplot(data=df, x='survived', ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Survival Status (Bar Plot)')
sns.boxplot(data=df, x='pclass', y='fare', ax=axes[1, 0], palette='Set2')
axes[1, 0].set_title('Fare by Class (Box Plot)')
sns.scatterplot(data=df, x='age', y='fare', hue='survived', ax=axes[1, 1])
axes[1, 1].set_title('Age vs Fare (Scatter Plot)')
plt.tight_layout()
plt.show()
```

```python
# 99. Create subplot using object-oriented syntax.
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
sns.histplot(df['age'].dropna(), ax=ax[0], color='teal', kde=True)
ax[0].set_title('Age Distribution')
ax[0].set_xlabel('Age')
sns.boxplot(data=df, x='pclass', y='fare', ax=ax[1], palette='Set2')
ax[1].set_title('Fare by Class')
ax[1].set_ylabel('Ticket Fare')
fig.suptitle('Titanic Data Analysis', fontsize=16)
plt.tight_layout()
plt.show()
```
