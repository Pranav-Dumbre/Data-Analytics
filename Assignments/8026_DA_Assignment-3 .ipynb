```python
# 24. Write a Pandas program convert the first and last character of each word to upper
# case in each word of a given series.
words = pd.Series(['php','python','java','dart'])
def capitalize_first_last(word):
    if len(word)==1:
        return word.upper()
    return word[0].upper()+word[1:-1]+word[-1].upper()

result = words.apply(capitalize_first_last)
print('original series:\n', words)
print('modified series:\n',result)
```

```python
# 28. Write a Pandas program to get the day of month, day of year, week number and day
# of week from a given series of date strings.
date_strings = pd.Series([
    '2026-04-01',
    '2026-04-02',
    '2026-04-03',
    '2026-04-04',
    '2026-04-05'
])
dates = pd.to_datetime(date_strings)

day_of_month = dates.dt.day
day_of_year = dates.dt.dayofyear
week_number = dates.dt.isocalendar().week
day_of_week = dates.dt.day_name()

result = pd.DataFrame({
    'Date': dates,
    'Day of Month': day_of_month,
    'Day of Year': day_of_year,
    'Week Number': week_number,
    'Day of Week': day_of_week
})
print(result)
```

```python
# 30. Write a Pandas program to filter words from a given series that contain atleast two
# vowels.
words = pd.Series(['Red', 'Green', 'Blue', 'Black', 'White', 'Sky', 'Cloud'])

def has_two_vowels(word):
    vowels = set("aeiouAEIOU")
    count = sum(1 for ch in word if ch in vowels)
    return count >= 2

filtered_words = words[words.apply(has_two_vowels)]
print(filtered_words)
```

```python
#31. Write a Pandas program to compute the Euclidean distance between two given series.
#Euclidean distance From Wikipedia,In mathematics, the Euclidean distance or Euclidean metric is the "ordinary" straight-line
#distance between two points in Euclidean space. With this distance, Euclidean space
#becomes a metric space. The associated norm is called the Euclidean norm.
series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series([1, 3, 5, 7, 9])

euclidean_distance = np.sqrt(((series1 - series2) ** 2).sum())
print('euclidean distance:',euclidean_distance)
```

```python
#33. Write a Pandas program to replace missing white spaces in a given string with the
# least frequent character.
from collections import Counter
s = pd.Series(['abc def ghi', 'hello world', 'pandas python'])

def replace_space_with_least_freq(text):
    freq = Counter(text.replace(" ", ""))
    least_freq_char = min(freq, key=freq.get)
    return text.replace(" ", least_freq_char)

result = s.apply(replace_space_with_least_freq)

print("Original Series:")
print(s)
print("\nModified Series:")
print(result)
```

```python
#34. Write a Pandas program to compute the autocorrelations of a given numeric series.
#From Wikipedia:Autocorrelation, also known as serial correlation, is the correlation of a signal with a
#delayed copy of itself as a function of delay. Informally, it is the similarity between
#observations as a function of the time lag between them.

s = pd.Series([1, 5, 3, 8, 10, 7, 6, 12])

autocorr_results = {lag: s.autocorr(lag=lag) for lag in range(1, 5)}
print("Original Series:")
print(s)
print("\nAutocorrelation values:")
for lag, value in autocorr_results.items():
    print(f"Lag {lag}: {value}")
```
