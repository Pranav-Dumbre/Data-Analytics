```python
print(arr**2)
print([x**2 for x in lst ])
```

```python
print(np.mean(arr))
print(sum(lst)/len(lst))
```

```python
#Array filtering
arr = np.arange(55,65)
print(arr)
```

```python
mask = arr>59
print("\ncondition (arr > 58):\n",mask)
```

```python
# step 4: filtered values
filtered =  arr[mask]
print("\nfiltered values (>59):\n", filtered)
```

```python
# Direct filtering
print(arr[arr>59])
```

```python
# multiple condition
print(arr[(arr>59) & (arr<63)])
```

```python
# students marks
marks = np.array([62,32,76,35,89])
passed_boolean = marks>35
print(passed_boolean)
print(marks[passed_boolean])
```

```python
# shallow copy vs Deep copy
a = np.array([10,20,30,40,50,60])
b= a[0:6]
b[0]=1000  # modify original array
```
