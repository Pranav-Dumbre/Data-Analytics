```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
# 1. line plot: average tips across the index
plt.figure(figsize=(6,4))
plt.plot(df.index,df['tip'],marker='o')
plt.title('Line plot- Tip across records')
plt.xlabel('Record index')
plt.ylabel('Tip')
plt.show();
```

```python
# line plot- seaborn
plt.figure(figsize=(6,4))
sns.lineplot(x=df.index,y=df['tip'])
plt.title('Line plot- Tip across records')
plt.xlabel('Record index')
plt.ylabel('Tip')
plt.show();
```

```python
# 2. Bar plot: Average total bill for each day
avg_bill_day = df.groupby('day')['total_bill'].mean()
# this will create new dataframe avg_bill_day, which will contain day and average bill
plt.figure(figsize=(6,4))
plt.bar(avg_bill_day.index,avg_bill_day.values)
plt.title('Bar plot- Average total bill by day')
plt.xlabel('day')
plt.ylabel('Average total bill')
plt.show();
```

```python
# bar plot - seaborn
plt.figure(figsize=(6,4))
sns.barplot(x='day',y='total_bill',data=df)
plt.title('Bar plot- Average total bill by day')
plt.xlabel('day')
plt.ylabel('Average total bill')
plt.show();
```

```python
#3. Histogram: Duistribution of bill
plt.figure(figsize=(6,4))
plt.hist(df['total_bill'],bins=15,edgecolor='black')
plt.title('Histogram-total bill')
plt.xlabel('Total bill')
plt.ylabel('frequency')
plt.show();
```

```python
# Histogram-seaborn
plt.figure(figsize=(6,4))
sns.histplot(df['total_bill'],bins=20,edgecolor='black')
plt.title('Histogram-total bill')
plt.xlabel('Total bill')
plt.ylabel('frequency')
plt.show();
```

```python
# 4.Box plot: Tip spread by day
tip_by_day = [df[df['day']==d]['tip'] for d in df['day'].unique()]
plt.figure(figsize=(6,4))
plt.boxplot(tip_by_day,labels=df['day'].unique())
plt.title('Boxplot- Tip by day')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show();
```

```python
# Box plot- seaborn
plt.figure(figsize=(6,4))
sns.boxplot(x='day',y='tip',data=df)
plt.title('Boxplot- Tip by day')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show();
```

```python
# 5. scatter plot: total bill vs tip
plt.figure(figsize=(6,4))
plt.scatter(df['total_bill'],df['tip'])
plt.title('scatter plot- Total bill vs Tip ')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show();
```

```python
# Scatter plot: seaborn
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.title('scatter plot- Total bill vs Tip ')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show();
```

```python
# Heatmap
corr = df[['total_bill', 'tip', 'size']].corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Seaborn HeatMap- Correlation mtrix')
plt.show()
```
