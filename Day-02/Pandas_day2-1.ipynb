```python
import numpy as np
import pandas as pd
```

```python
# Numpy arrays
company = np.array(['TCS','Infosys','Wipro'])
sales = np.array([100,200,300])
profit = np.array([20,50,70])
# Access profit for TCS
print('Company: ',company)
print('Sales: ',sales)
print('Profit: ',profit)

#accessing elements 
print("Numpy Array")
print("Infosys sales:",sales[1])
print("Infosys profit:",profit[1])
```

```python
# Pandas series
sales_series=pd.Series(sales,index=company)
profit_series=pd.Series(profit,index=company)
print(sales_series)
print(profit_series)

#accessing using label
print("\naccessing using label\n")
print("Infosys sales:",sales_series['Infosys'])
print("Infosys profit:",profit_series['Infosys'])
```

```python
prn = np.array([8003,8026,8044])
name = np.array(["Ajinkya","Pranav","Shubham"])
mobile = np.array([9898435,998745,125234])
name_series=pd.Series(name,index=prn)
mobile_series=pd.Series(mobile,index=prn)
print("\n Third student name \n")
print("Third student name: name_series[prn[2]] :",name_series[prn[2]])
print("Third student name: name_series[8044] :",name_series[8044])
print("Third student name: name_series.loc[8044] :",name_series.loc[8044])
print("Third student name: name_series.iloc[2] :",name_series.iloc[2])

print("\n second student mobile number \n")
print("Second student mobile: mobile_series[prn[1]] :",mobile_series[prn[1]])
print("Second student mobile: mobile_series[8026] :",mobile_series[8026])
print("Second student mobile: mobile_series.loc[8026] :",mobile_series.loc[8026])
print("Second student mobile: mobile_series.iloc[1] :",mobile_series.iloc[1])
```

```python
# Dataframe
df = pd.DataFrame({
    'Sales': sales,
    'Profit': profit
},index=company)
df
```

```python
# add new column
df['Margin']=df['Profit']/df['Sales']*100
df
```

```python
# filter multiple column
#print("\nHigh profit companies:\n",df[df['Profit']>40])
df[df['Profit']>40]
```

```python
# combined condition
df[(df['Sales']>150)&(df['Profit']>40)]
```

```python
# sample data
df = pd.DataFrame({
    'Employee':['Pranav','Ajinkya','Shubham','Chirayu','Rudraksh'],
    'Department':['CS','IT','IT','HR','Finace'],
    'Salary':[50000,60000,np.nan,65000,70000]
})
df
```

```python
# indexing
df_indexed = df.set_index('Employee')
df_indexed
```

```python
# filtering
df[df['Salary']>60000]
```

```python
# Check missing
df.isnull()
```

```python
# Drop missing rows
df.dropna()
```

```python
# Grouping
df.groupby('Department')['Salary'].mean()
```

```python
# merge
df_bonus= pd.DataFrame({
    'Employee':['Ajinkya','Pranav','Chirayu'],
    'Bonus':[5000,7000,4000]
})
pd.merge(df,df_bonus, on='Employee')
```

```python
# concat
df_new = pd.DataFrame({
    'Employee':['Rajveer'],
    'Department':['CS'],
    'Salary':[55000]
})
pd.concat([df,df_new])
```
